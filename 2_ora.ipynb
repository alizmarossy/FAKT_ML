```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
```

```python
file_url = "http://storage.googleapis.com/download.tensorflow.org/data/heart.csv"
```

```python
dataframe = pd.read_csv(file_url)
```

```python
y_df = dataframe.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_df, y_df, test_size=0.1, shuffle=True)
# randomizálja a splitet
```

#Architektúra létrehozása

```python
from tensorflow.keras.layers import Input, Dense
import keras
from tensorflow.keras.utils import plot_model
from tensorflow.keras.metrics import BinaryAccuracy
```

```python
input = Input(shape = (x_train.shape[1]))
x = Dense(32, activation = 'relu')(input)
x = Dense(64, activation = 'relu')(x)
x = Dense(64, activation = 'relu')(x)
output = Dense(1, activation= 'sigmoid')(x)
#1. rejtett, Dense réteg, csak előrefele halad az információ
#minden neuron mindegyikkel össze van kötve
#2 hatványok
# output 1 neuron
#újratanításnál ezt a cellát is le kell futtatni
```

```python
model = keras.Model(input, output)
```

```python
plot_model(model, show_shapes = True)
```

```python
model.compile(loss = 'binary_crossentropy', metrics = [BinaryAccuracy()])
```

```python
model.fit(x = x_train, y = y_train, epochs = 30)
#epochs : hányszor ment át rajta az adat
```

```python
model.predict(x_test)
```

```python
model.evaluate(x = x_test, y = y_test)
```
