```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
```

#ArchitektÃºra lÃ©trehozÃ¡sa

```python
from tensorflow.keras.layers import Input, Dense
import keras
from tensorflow.keras.utils import plot_model
from tensorflow.keras.metrics import BinaryAccuracy
```

```python
input = Input(shape = (x_train.shape[1]))
x = Dense(32, activation = 'relu')(input)
x = Dense(64, activation = 'relu')(x)
x = Dense(64, activation = 'relu')(x)
output = Dense(1, activation= 'sigmoid')(x)
#1. rejtett, Dense rÃ©teg, csak elÅ‘refele halad az informÃ¡ciÃ³
#minden neuron mindegyikkel Ã¶ssze van kÃ¶tve
#2 hatvÃ¡nyok
# output 1 neuron
#ÃºjratanÃ­tÃ¡snÃ¡l ezt a cellÃ¡t is le kell futtatni
```

```python
model = keras.Model(input, output)
```

```python
plot_model(model, show_shapes = True)
```

```python
model.compile(loss = 'binary_crossentropy', metrics = [BinaryAccuracy()])
```

```python
model.fit(x = x_train, y = y_train, epochs = 30)
#epochs : hÃ¡nyszor ment Ã¡t rajta az adat
```

```python
model.predict(x_test)
```

```python
model.evaluate(x = x_test, y = y_test)
```
