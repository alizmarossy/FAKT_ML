```python
rng = np.random.default_rng(255)
miez = rng.random(59885568)
kiez = miez.reshape(3648,5472,3)
```

```python
kiez = kiez*100
print(kiez)
```

```python
kiez_piros = kiez.copy()
kiez_piros[:,:,(1,2)] = 0
plt.imshow(kiez_piros)
```

```python
ezmi = rng.integers(255, size=(3648, 5472,3))
ezmi.shape
plt.imshow(ezmi)
#így sem látszódnak a színek
```

```python
ezmi2 = rng.integers(255, size=(100, 100,3))
ezmi2.shape
plt.imshow(ezmi2)
#nagyobb pixelek, hogy látszódjanak a színek
```

##RGB

```python
ezmi2_piros = ezmi2.copy()
ezmi2_piros[:,:,(1,2)] = 0
plt.imshow(ezmi2_piros)
```

```python
ezmi2_zold = ezmi2.copy()
ezmi2_zold[:,:,(0,2)] = 0
plt.imshow(ezmi2_zold)
```

```python
ezmi2_kek = ezmi2.copy()
ezmi2_kek[:,:,(0,1)] = 0
plt.imshow(ezmi2_kek)
```

```python
import cv2
fig = plt.figure(figsize=(10, 7))
rows = 1
columns = 3

fig.add_subplot(rows, columns, 1)
plt.imshow(ezmi2_piros)
fig.add_subplot(rows, columns, 2)
plt.imshow(ezmi2_zold)
fig.add_subplot(rows, columns, 3)
plt.imshow(ezmi2_kek)
```
