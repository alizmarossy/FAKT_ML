#AlexNet elkészítése

##CIFAR10 adatbázis importálása

```python
from tensorflow import keras
import matplotlib.pyplot as plt
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.layers import Conv2D, Input, MaxPooling2D
```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.cifar10.load_data()
```

```python
x_train[0].shape
```

```python
class_names = ['repulo', 'auto', 'madar', 'macska', 'szarvas', 'kutya', 'beka', 'lo', 'hajo', 'kamion']
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,20))
for i in range(30):
    image = x_train[i]
    label = y_train[i]
    ax = plt.subplot(5,5,i+1)
    plt.title(class_names[label[0]])
    plt.imshow(image)
    plt.axis('off')
```

##Adatok formálása

```python
x_train = x_train.astype('float32') / 255
x_test = x_test.astype('float32') / 255
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
y_train = to_categorical(y_train, num_classes = len(class_names))
y_test = to_categorical(y_test, num_classes = len(class_names))
```

##AlexNet elkészítése

```python
from tensorflow.keras.layers import Conv2D, Input, MaxPooling2D, Dense, Flatten, Dropout
```

```python
input = Input(shape= (32,32,3))
x = Conv2D(filters=64, kernel_size=5, strides = 4, activation = 'relu', padding = 'same')(input)
x = MaxPooling2D(pool_size=2, strides=2)(x)
x = Conv2D(filters=256, kernel_size=5, strides = 4, activation = 'relu', padding = 'same')(x)
x = MaxPooling2D(pool_size=2, strides=2)(x)
x = Conv2D(filters=384, kernel_size=3, strides = 4, activation = 'relu', padding = 'same')(x)
x = Conv2D(filters=384, kernel_size=3, strides = 4, activation = 'relu', padding = 'same')(x)
x = Conv2D(filters=192, kernel_size=3, strides = 4, activation = 'relu', padding = 'same')(x)
x = MaxPooling2D(pool_size=2, strides=2)(x)
x = Flatten()(x)

x = Dense(4096, activation = 'relu')(x)
x = Dropout(0.2)(x)
x = Dense(4096, activation = 'relu')(x)
x = Dropout(0.4)(x)

output = Dense(len(class_names), activation='softmax')(x)
```

```python
AlexNet = keras.Model(input, output)
```
